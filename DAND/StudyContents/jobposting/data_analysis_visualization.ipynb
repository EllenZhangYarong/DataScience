```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./dataset/clean-job-postings.csv')
```

```python
df.head()
```

```python
# Number of 'ASAP' start dates (numerator)
asap_counts = df.StartDate.value_counts()['ASAP']
asap_counts
```

```python
# Number of non-empty start dates (denominator)
non_empty_counts = sum(df.StartDate.value_counts())
non_empty_counts
```

```python
non_empty_counts2 = df.StartDate.count()
non_empty_counts2
```

```python
# Percentage of positions with an urgent start date i.e. 'ASAP'
asap_counts / non_empty_counts
```
